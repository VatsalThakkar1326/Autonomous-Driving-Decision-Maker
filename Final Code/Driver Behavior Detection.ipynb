```python
import cv2
import numpy as np
from tensorflow.keras.models import load_model

# Function for preprocessing the frame before making predictions
def preprocess_frame(frame):
    # Resize the frame to match the input size of your model
    processed_frame = cv2.resize(frame, (240, 240))  # Adjust the size as needed
    # Normalize pixel values to the range [0, 1]
    processed_frame = processed_frame / 255.0
    return processed_frame

# Load your deep learning model
# Replace 'your_model_path.h5' with the path to your model file
model = load_model("C:\\Users\\14374\\Downloads\\your_model.h5")
class_labels = ["Other", "Safe Driving", "Talking Phone", "Texting Phone", "Turning"]

# Open the camera (use 0 for the default camera)
cap = cv2.VideoCapture(1)

# Open a text file for writing predictions
output_file = open("predictions.txt", "w")
line_counter = 0

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # Preprocess the frame for your deep learning model
    processed_frame = preprocess_frame(frame)

    # Make predictions using your deep learning model
    predictions = model.predict(np.expand_dims(processed_frame, axis=0), verbose=0)

    # Get the predicted class label
    predicted_class_index = np.argmax(predictions)
    predicted_class_label = class_labels[predicted_class_index]
    print("Predicted Class:", predicted_class_label)

    # Write the predicted class label to the output file
    output_file.write(predicted_class_label + "\n")
    line_counter += 1

    # Flush the buffer to write in real-time
    output_file.flush()

    # Check if line_counter reaches 50, then clear the file and reset the counter
    if line_counter == 50:
        output_file.close()
        output_file = open("predictions.txt", "w")
        line_counter = 0

    # Display the original frame with predicted class label
    cv2.putText(frame, predicted_class_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    cv2.imshow('Camera Feed', frame)

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the camera, close the output file, and close all OpenCV windows
cap.release()
output_file.close()
cv2.destroyAllWindows()
```
